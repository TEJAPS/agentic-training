```python
# Self-Attention Visualization (toy example in Python):
import numpy as np

# Example sentence: "The dog chased the ball"
tokens = ["The", "dog", "chased", "the", "ball"]

# Fake attention weights (just for demo)
attention = np.array([
    [0.1, 0.3, 0.2, 0.2, 0.2],  # "The"
    [0.2, 0.1, 0.4, 0.2, 0.1],  # "dog"
    [0.1, 0.4, 0.1, 0.3, 0.1],  # "chased"
    [0.2, 0.2, 0.2, 0.1, 0.3],  # "the"
    [0.1, 0.2, 0.3, 0.2, 0.2]   # "ball"
])

# Display attention of "dog"
print("Attention for 'dog':")
for token, score in zip(tokens, attention[1]):
    print(f"{token}: {score:.2f}")

#Shows how “dog” attends more to “chased” than to “ball.”
```
